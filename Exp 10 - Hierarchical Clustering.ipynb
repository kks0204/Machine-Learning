```python
import numpy as np                
import pandas as pd               
import seaborn as sns             
import matplotlib.pyplot as plt   
import scipy.stats                
from sklearn import preprocessing
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = pd.read_csv("..\Downloads\Iris.csv")
iris.head()
```

```python
g = sns.PairGrid(iris.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]])
g.map_diag(plt.hist, histtype="step", linewidth=3)
g.map_offdiag(plt.scatter)
```

```python
# single linkage
dist_sin = linkage(iris.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]],method="single")
plt.figure(figsize=(18,6))
dendrogram(dist_sin, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("DENDROGRAM SINGLE LINKAGE METHOD",fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(24,4))

plt.suptitle("Hierarchical Clustering Single Linkage Method",fontsize=18)

plt.subplot(1,3,3)
plt.title("Species",fontsize=14)
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm", data=iris, hue="Species")
```

```python
dist_comp = linkage(iris.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]],method="complete")



plt.figure(figsize=(18,6))
dendrogram(dist_comp, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("DENDROGRAM COMPLETE LINKAGE METHOD",fontsize=18) 
plt.show()
```

```python
plt.figure(figsize=(24,4))

plt.suptitle("Hierarchical Clustering Complete linkage Method",fontsize=18)

plt.subplot(1,3,3)
plt.title("Species",fontsize=14)
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm", data=iris, hue="Species")
```

```python
g = sns.PairGrid(iris, vars=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"], hue='Species')
g.map(plt.scatter)
g.add_legend()
```
